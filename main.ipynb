```python
# import libraries
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score

import warnings
warnings.filterwarnings('ignore')
```

```python
# Draw S&P500 index data and query them

sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period="max")
```

```python
# Display data

sp500
```

```python
# Changing index to desired format
sp500.index = pd.to_datetime(sp500.index.date)
sp500 
```

```python
# Plot the data on closing price
sp500.plot.line(y = 'Close')
```

```python
# Retaining the data requied and dropping the others
sp500 = sp500[['Open', 'High', 'Low', 'Close', 'Volume']]
sp500
```

```python
# Create column for the next day

sp500['Next Day'] = sp500['Close'].shift(-1)
sp500.head()
```

```python
# Set target value for machine learning

sp500['Target'] = np.where(sp500['Next Day'] > sp500['Close'], 1, 0)
sp500
```

```python
# Slice data since 1980 till present
sp500 = sp500.loc['1980-01-01':].copy()
```

```python
# Train the machine learning model with test and train data

model = RandomForestClassifier(n_estimators = 200, random_state = 1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]


predictors = ['Open', 'High', 'Low', 'Close', 'Volume']
model.fit(train[predictors], train['Target'] )
```

```python
# Make predicctions with the model
pred = model.predict(test[predictors])
pred = pd.Series(pred, index = test.index)
pred
```

```python
# Calculate precision score
precision_score(test['Target'], pred)
```

```python
# Combine the target and prediction of test data for plotting

test_pred_target = pd.concat([test['Target'], pred], axis=1)

test_pred_target.plot()
```

```python
# Create function to make predictions with machine learning models

def prediction(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    pred = model.predict(test[predictors])
    pred = pd.Series(pred, index = test.index, name = 'pred')
    combined_result = pd.concat([test['Target'], pred], axis = 1)
    return combined_result
```

```python
# Create function to backtest the model

def backtest(data, model, predictors, start = 2500, step = 250):
    predictions_df = []
    
    for i in range(2500, data.shape[0], 250):
        X = data.iloc[0:i].copy()
        y = data.iloc[i:(i+step)].copy()
        pred = prediction(X, y, predictors, model)
        predictions_df.append(pred)
        
    return pd.concat(predictions_df)
```

```python
# Conduct backtesting

predictions = backtest(sp500, model, predictors)
```

```python
# Obtain the precision score

precision_score(predictions['Target'], predictions['pred'])
```

```python
# Adding predictors to increase the accuracy of the model

horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    
    rolling = sp500.rolling(horizon).mean()
    
    ratio = f"Close_Ratio {horizon}"
    sp500[ratio] = rolling['Close'] / sp500['Close']
    
    trend = f"Trend {horizon}"
    sp500[trend] = sp500.shift(1).rolling(horizon).sum()['Target']
    
    new_predictors += [ratio, trend]
```

```python
# Remove rows of sp500 with NaNs for proper machine learning

sp500.dropna(inplace = True)
sp500
```

```python
# Enhancing the model

model = RandomForestClassifier(n_estimators = 300, min_samples_split = 50, random_state = 1)
```

```python
# Create function with new predictors

def new_predict(test, train, predictors, model):
    model.fit(train[predictors], train['Target'])
    pred = model.predict_proba(test[predictors])
    pred = np.where(pred >= .6, 1, 0)
    pred = pd.Series(pred, index = test.index, name = 'pred')
    combined = pd.concat([test['Target'], pred], axis=1)
    
    return combined
```

```python
# Backtest the model with the new predictors

pred_new = backtest(sp500, model, new_predictors)
```

```python
# Obtain the precision score with the new predictors

precision_score(pred_new['Target'], pred_new['pred'])
```
