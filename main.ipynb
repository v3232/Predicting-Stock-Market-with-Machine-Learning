```python
# import libraries
import yfinance as yf
```

```python
# Draw S&P500 index data and query them

sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period="max")
```

```python
# Display data

sp500
```

```python
# Changing index to desired format
sp500.index = sp500.index.date
sp500
```

```python
# Plot the data
sp500.plot.line(y = 'Close')
```

```python
# Retaining the data requied and dropping the others
sp500 = sp500[['Open', 'High', 'Low', 'Close', 'Volume']]
sp500
```
