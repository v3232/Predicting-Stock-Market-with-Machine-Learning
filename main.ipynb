```python
# import libraries
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
# Draw S&P500 index data and query them

sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period="max")
```

```python
# Display data

sp500
```

```python
# Changing index to desired format
sp500.index = pd.to_datetime(sp500.index.date)
sp500 
```

```python
# Plot the data on closing price
sp500.plot.line(y = 'Close')
```

```python
# Retaining the data requied and dropping the others
sp500 = sp500[['Open', 'High', 'Low', 'Close', 'Volume']]
sp500
```

```python
# Create column for the next day

sp500['Next Day'] = sp500['Close'].shift(-1)
sp500.head()
```

```python
# Set target value for machine learning


# sp500['Target'] = (sp500['Next Day'] > sp500['Close']).astype(int)
sp500['Target'] = np.where(sp500['Next Day'] > sp500['Close'], 1, 0)
sp500
```

```python
# Slice data since 1980 till present
sp500 = sp500.loc['1980-01-01':].copy()
sp500
```
